```python
def compute_cosine_similarity(user_item_matrix):
    """
    Computes cosine similarity for all user vectors in a user-item matrix, ignoring zero values.

    Parameters:
    user_item_matrix (numpy.ndarray): A 2D numpy array where rows represent users and columns represent items.

    Returns:
    numpy.ndarray: A 2D numpy array containing cosine similarity scores between users.
    """
    # Mask zero values to ignore them in the similarity calculation
    masked_matrix = np.ma.masked_equal(user_item_matrix, 0)
    
    # Number of users
    num_users = user_item_matrix.shape[0]
    
    # Initialize the similarity matrix
    similarity_matrix = np.zeros((num_users, num_users))
    
    # Compute cosine similarity while ignoring zero values
    for i in range(num_users):
        for j in range(num_users):
            if i == j:
                similarity_matrix[i, j] = 1.0
            else:
                u = masked_matrix[i]
                v = masked_matrix[j]
                
                # Calculate the dot product and magnitudes while ignoring masked values
                dot_product = np.sum(u * v)
                magnitude_u = np.sqrt(np.sum(u * u))
                magnitude_v = np.sqrt(np.sum(v * v))
                
                if magnitude_u == 0 or magnitude_v == 0:
                    similarity_matrix[i, j] = 0.0
                else:
                    similarity_matrix[i, j] = dot_product / (magnitude_u * magnitude_v)
    
    return similarity_matrix

```

```python
def compute_item_cosine_similarity(user_item_matrix):
    """
    Computes cosine similarity for all item vectors in a user-item matrix, ignoring zero values.

    Parameters:
    user_item_matrix (numpy.ndarray): A 2D numpy array where rows represent users and columns represent items.

    Returns:
    numpy.ndarray: A 2D numpy array containing cosine similarity scores between items.
    """
    # Mask zero values to ignore them in the similarity calculation
    masked_matrix = np.ma.masked_equal(user_item_matrix.T, 0)
    
    # Number of items
    num_items = user_item_matrix.shape[1]
    
    # Initialize the similarity matrix
    similarity_matrix = np.zeros((num_items, num_items))
    
    # Compute cosine similarity while ignoring zero values
    for i in range(num_items):
        for j in range(num_items):
            if i == j:
                similarity_matrix[i, j] = 1.0
            else:
                u = masked_matrix[i]
                v = masked_matrix[j]
                
                # Calculate the dot product and magnitudes while ignoring masked values
                dot_product = np.sum(u * v)
                magnitude_u = np.sqrt(np.sum(u * u))
                magnitude_v = np.sqrt(np.sum(v * v))
                
                if magnitude_u == 0 or magnitude_v == 0:
                    similarity_matrix[i, j] = 0.0
                else:
                    similarity_matrix[i, j] = dot_product / (magnitude_u * magnitude_v)
    
    return similarity_matrix
```

```python
def predict_rating(user_item_matrix, user_index, item_index, similarity_matrix, max_class, top_n=10):
    """
    Predicts the rating for a given user-item pair using collaborative filtering.

    Parameters:
    user_item_matrix (numpy.ndarray): A 2D numpy array where rows represent users and columns represent items.
    user_index (int): The index of the user for whom the rating is to be predicted.
    item_index (int): The index of the item for which the rating is to be predicted.
    similarity_matrix (numpy.ndarray): A 2D numpy array containing cosine similarity scores between users.
    top_n (int): The number of top similar users to consider for the prediction.

    Returns:
    float: The predicted rating.
    """
    user_ratings = user_item_matrix[user_index]
    user_mean_rating = np.mean(user_ratings[user_ratings > 0])
    
    # Get similarities and sort by highest similarity
    similarities = similarity_matrix[user_index]
    similar_users = np.argsort(-similarities)  # sort in descending order of similarity

    
    numerator = 0
    denominator = 0
    count = 0
    
    for similar_user in similar_users:
        if count >= top_n:
            break
        
        if user_item_matrix[similar_user, item_index] > 0 and similar_user != user_index:
            similar_user_ratings = user_item_matrix[similar_user]
            similar_user_mean_rating = np.mean(similar_user_ratings[similar_user_ratings > 0])
            similarity = similarities[similar_user]
            
            numerator += similarity * (user_item_matrix[similar_user, item_index] - similar_user_mean_rating)
            denominator += np.abs(similarity)
            count += 1
    
    if denominator == 0:
        return user_mean_rating
    
    predicted_rating = user_mean_rating + (numerator / denominator)
    
    return predicted_rating 
```

```python
def predict_rating_item(user_item_matrix, user_index, item_index, similarity_matrix, top_n=10):
    """
    Predicts the rating for a given user-item pair using item-item collaborative filtering.

    Parameters:
    user_item_matrix (numpy.ndarray): A 2D numpy array where rows represent users and columns represent items.
    user_index (int): The index of the user for whom the rating is to be predicted.
    item_index (int): The index of the item for which the rating is to be predicted.
    similarity_matrix (numpy.ndarray): A 2D numpy array containing cosine similarity scores between items.
    top_n (int): The number of top similar items to consider for the prediction.

    Returns:
    float: The predicted rating.
    """
    user_ratings = user_item_matrix[user_index]
    
    # Get similarities and sort by highest similarity
    similarities = similarity_matrix[item_index]
    similar_items = np.argsort(-similarities)  # sort in descending order of similarity

    
    numerator = 0
    denominator = 0
    count = 0
    
    for similar_item in similar_items:
        if count >= top_n:
            break
        
        if user_ratings[similar_item] > 0 and similar_item != item_index:
            similarity = similarities[similar_item]
            
            numerator += similarity * user_ratings[similar_item]
            denominator += np.abs(similarity)
            count += 1
    
    if denominator == 0:
        return 0  # If no similar items with ratings, return 0 or some other default value
    
    predicted_rating = numerator / denominator
    
    return predicted_rating
```
