```python
n_classes = 2
n_items = m_base.shape[1]
n_votes = 5
class_prob = np.full(n_classes, 1/n_classes)
vote_prob = np.full((n_classes, n_items, n_votes), 1/(n_votes))
n_users = m_base.shape[0] 
```

```python
def e_step(data, class_prob, vote_prob):
        
    n_users, n_items = data.shape

    n_classes = class_prob.shape[0]

    responsibilities = np.zeros((n_users, n_classes))  # Î³_uc
        
    for u in range(n_users):
        for c in range(n_classes):
            prod = 1
            for i in range(n_items):
                if data[u, i] != -1:
                    prod *= vote_prob[c, i, data[u, i] - 1]
                    if u == 404:
                        print(f"{prod=} {i=}")
            responsibilities[u, c] = class_prob[c] * prod
        responsibilities[u, :] /= np.sum(responsibilities[u, :])  # Normalize responsibilities
        
    return responsibilities
```

```python
def q_function(params, data, responsibilities, n_classes, n_votes, n_users, n_items):

    #n_users, n_items = data.shape

    class_prob = params[:n_classes]

    vote_prob = params[n_classes:].reshape(n_classes, n_items, n_votes)

    q_value = 0
        
    for u in range(n_users):
        for c in range(n_classes):
            log_prob = np.log(class_prob[c])
            for i in range(n_items):
                if data[u, i] != -1:
                    log_prob += np.log(vote_prob[c, i, data[u, i] - 1])
            print(f'resp[{u}, {c}] = {responsibilities[u, c]}')
            q_value += responsibilities[u, c] * log_prob
            print(f'q_value = {q_value}')
        
    return -q_value  # We minimize, so return negative
```

```python
def m_step(data, responsibilities, class_prob, vote_prob, n_votes):

    n_items = data.shape[1]

    n_classes = class_prob.shape[0]
        
    initial_params = np.concatenate([class_prob, vote_prob.flatten()])
        
    bounds = ((0, 1)) * len(initial_params)

        
    result = minimize(q_function, x0 = initial_params, args=(data, responsibilities, n_classes, n_votes), bounds=bounds)
        
    print('Ciao')


    optimized_params = result.x
    class_prob = optimized_params[:n_classes]
    vote_prob = optimized_params[n_classes:].reshape(n_classes, n_items, n_votes)
    return class_prob, vote_prob
```

```python
def fit(data, class_prob, vote_prob, max_iter=100):
    log_likelihoods = []
    c_p = class_prob.copy()
    v_p = vote_prob.copy()
    for _ in range(max_iter):
        responsibilities = e_step(data, c_p, v_p)  # E-step
        c_p, v_p = m_step(data, responsibilities, c_p, v_p)  # M-step
        log_l = log_likelihood(data, c_p, v_p)
        log_likelihoods.append(log_l)
    return c_p, v_p, log_likelihoods
```

```python
def log_likelihood(data, class_prob, vote_prob):

    n_users, n_items = data.shape

    n_classes = class_prob.shape[0]
        
    for u in range(n_users):
        user_likelihood = 0
        for c in range(n_classes):
            class_likelihood = class_prob[c]
            for i in range(n_items):
                if data[u, i] != -1:
                    class_likelihood *= vote_prob[c, i, data[u, i] - 1]
            user_likelihood += class_likelihood
        log_likelihood += np.log(user_likelihood)
        
    return log_likelihood
```
